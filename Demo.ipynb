```python
from extractor import TableExtractorOCR

import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import cv2
```

```python
def display_image(image, figsize=(15,10)):
    fig = plt.figure(figsize=figsize)
    ax = fig.add_subplot(111)
    ax.imshow(image)
```

```python
image = cv2.imread('Images/grid_table01.png')
display_image(image)
```

```python
ex = TableExtractorOCR()
found, tables = ex.get_tables('Images/grid_table01.png')

if found:
    print('No of Tables :', len(tables))
    for table in tables:
        data = ex.get_text_from_table(table)
        df = pd.DataFrame.from_dict(data).T
else: 
    print('No Tables detected')
df
```
