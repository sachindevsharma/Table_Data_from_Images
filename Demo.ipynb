```python
import cv2
import matplotlib.pyplot as plt
from table_extractor import TableExtractor
```

```python
def display_image(image, figsize=(15,10)):
    try:
        img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    except: 
        pass
    fig = plt.figure(figsize=figsize)
    ax = fig.add_subplot(111)
    ax.imshow(image, cmap='gray')
```

```python
ex = TableExtractor()
```

## Table Detection

```python
image = cv2.imread("Table Detection/data/Report_pics_apple/out_29.png")
display_image(image, (15,20))
```

```python
tables, bordered_tables, data = ex.find_tables(image)

for table in tables:
    display_image(table)
    
for table in bordered_tables:
    display_image(table)
```
